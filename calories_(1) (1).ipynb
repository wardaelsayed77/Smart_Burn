```python

num_features = ['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp', 'Calories']

def detect_outliers(df, features):
    outlier_dict = {}
    for col in features:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        outlier_dict[col] = outliers
        print(f"{col}: {len(outliers)} outliers")

    return outlier_dict

outliers = detect_outliers(dataset_reg, num_features)
```

```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

for col in ['Height', 'Weight', 'Heart_Rate', 'Calories','Body_Temp']:
    dataset_reg = remove_outliers(dataset_reg, col)
```

```python
from sklearn.ensemble import StackingRegressor
base_models = [
    ('dt', DecisionTreeRegressor()),
    ('knn', KNeighborsRegressor(n_neighbors=5)),
    ('rf', RandomForestRegressor(n_estimators=100))
]

final_model = LinearRegression()

stack = StackingRegressor(
    estimators=base_models,
    final_estimator=final_model
)

stack.fit(X_train, y_train)
y_train_pred = stack.predict(X_train)
r2_train = r2_score(y_train, y_train_pred)
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
mae_train = mean_absolute_error(y_train, y_train_pred)

y_test_pred = stack.predict(X_test)
r2_test = r2_score(y_test, y_test_pred)
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
mae_test = mean_absolute_error(y_test, y_test_pred)

print("Train R2:", r2_train, " | Train RMSE:", rmse_train , " | Train MAE:", mae_train)
print("Test R2: ", r2_test, " | Test RMSE: ", rmse_test , " | Test MAE:", mae_test)
```

```python

from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline


kernels = ['linear', 'rbf', 'poly', 'sigmoid']

results = []

for kernel in kernels:
    if kernel == 'poly':
        model = make_pipeline(StandardScaler(), SVR(kernel=kernel, degree=3, C=100, epsilon=0.1))
    elif kernel == 'rbf' or kernel == 'sigmoid':
        model = make_pipeline(StandardScaler(), SVR(kernel=kernel, C=100, gamma=0.1, epsilon=0.1))
    else:
        model = make_pipeline(StandardScaler(), SVR(kernel=kernel, C=100, epsilon=0.1))

    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    r2_train = r2_score(y_train, y_train_pred)
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    mae_train = mean_absolute_error(y_train, y_train_pred)

    y_test_pred = model.predict(X_test)
    r2_test = r2_score(y_test, y_test_pred)
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    mae_test = mean_absolute_error(y_test, y_test_pred)

    results.append({
        'kernel': kernel,
        'r2_train': r2_train,
        'rmse_train': rmse_train,
        'r2_test': r2_test,
        'rmse_test': rmse_test,
        'mae_train': mae_train,
        'mae_test': mae_test
    })

results_df = pd.DataFrame(results)
print(results_df)
```
